```python
# Get the venues
venues = result['response']['groups'][0]['items']
dataframe = json_normalize(venues)
```

```python
# Extract the necessary info and form a new dataframe
foodinfo=dataframe[['venue.id','venue.name', 'venue.location.address', 'venue.location.lat', 'venue.location.lng','venue.delivery.provider.name']]
foodinfo.columns=['ID','Chinese Restaurant','Address','Lat','Lng','Delivery Provider ']

foodinfo.head()
```

### Get the info of each restaurant

```python
r=[]
for vid in foodinfo.ID: 
    url1 = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(vid, CLIENT_ID, CLIENT_SECRET, VERSION)
    resultall=requests.get(url1).json()
    try:
        r.append(resultall['response']['venue']['rating'])
    except:
        r.append('Not Rated')
foodinfo['Rating']= r
```

```python
# find the new york map 
address = 'New York City, NY'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))

map_newyork = folium.Map(location=[latitude,longitude], zoom_start=13)


# map each restaurant
for lat, lng, name,rat in zip(foodinfo['Lat'], foodinfo['Lng'],foodinfo['Chinese Restaurant'],foodinfo['Rating']):
    label = '{},{}'.format(name, rat)
    label = folium.Popup(label, parse_html=True)
    rat = folium.Popup(rat,parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
